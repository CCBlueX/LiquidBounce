```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv("jumpLengths.csv")

data
```

```python
fit = np.polyfit(data['Fall Distance'].values[1:], data['Jump Length'].values[1:], deg=2)

xp = np.linspace(-1, 10, 100)
poly = np.poly1d(fit)

def approxFunction(x):
    if x <= 0:
        return 2.7
    else:
        return poly(x) - 0.1

plt.plot(data['Fall Distance'], data['Jump Length'])
plt.plot(xp, list(map(approxFunction, xp)), linestyle='--')

plt.xlabel("Fall distance")
plt.ylabel("Jump length")

plt.show()

print(fit)
```
